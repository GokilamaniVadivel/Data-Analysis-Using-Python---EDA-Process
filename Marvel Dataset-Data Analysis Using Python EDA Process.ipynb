## Retrieving data from a data frame

The first thing you might want to do is retrieve data from this data frame, e.g., the counts of a specific day or the list of values in a particular column. To do this, it might help to understand the internal representation of data in a data frame. Conceptually, you can think of a dataframe as a dictionary of lists: keys are column names, and values are lists/arrays containing data for the respective columns.

```python
marvel_data_dict={
    'Title': ['Howard the Duck', 'Blade', 'Blade II', 'Blade: Trinity', 'Ghost Rider', 'Ghost Rider: Spirit of Vengeance'],
    'worldwide_gross': [37897665, 131183530, 155010032, 128905366, 228738393, 132563930],
    'distribution': ['Universal Pictures', 'New Line Cinema', 'New Line Cinema', 'New Line Cinema', 'Columbia Pictures', 'Columbia Pictures'],
}   
```

With the dictionary of lists analogy in mind, you can now guess how to retrieve data from a data frame. For example, we can get a list of values from a specific column using the `[]` indexing notation.

```python
marvel_data_dict['Title']
```

```python
marvel_data_dict['worldwide_gross']
```

```python
marvel_data_dict['distribution']
```

```python
marvel_data['Title']
```

Pandas also provides the `.at` method to retrieve the element at a specific row & column directly.

```python
marvel_data.at[35, 'Title']
```

Instead of using the indexing notation `[]`, Pandas also allows accessing columns as properties of the dataframe using the `.` notation. However, this method only works for columns whose names do not contain spaces or special characters.

```python
marvel_data.Title
```

Further, you can also pass a list of columns within the indexing notation `[]` to access a subset of the data frame with just the given columns.

```python
marvel = marvel_data[['Title', 'Distributor(s)']]
marvel
```

The new data frame `marvel` is simply a "view" of the original data frame `marvel_data`. Both point to the same data in the computer's memory. Changing any values inside one of them will also change the respective values in the other. Sharing data between data frames makes data manipulation in Pandas blazing fast. You needn't worry about the overhead of copying thousands or millions of rows every time you want to create a new data frame by operating on an existing one.

Sometimes you might need a full copy of the data frame, in which case you can use the `copy` method.

```python
marvel_data_copy = marvel_data.copy()

marvel_data_copy
```

The data within `marvel_data_copy` is completely separate from `marvel_data`, and changing values inside one of them will not affect the other.

## Bonus: Basic Plotting with Pandas

We generally use a library like `matplotlib` or `seaborn` plot graphs within a Jupyter notebook. However, Pandas dataframes & series provide a handy `.plot` method for quick and easy plotting.

Let's plot a line graph showing how the number of daily cases varies over time.

```python
import matplotlib.pyplot as plt

# Sorting data based on Worldwide Revenue
marvel_data_sorted = marvel_data.sort_values(by='Worldwide', ascending=False).head(10)

plt.figure(figsize=(12, 6))
plt.barh(marvel_data_sorted['Title'], marvel_data_sorted['Worldwide'], color='skyblue')
plt.xlabel("Revenue (in billions)")
plt.ylabel("Movies")
plt.title("Top 10 Marvel Movies by Worldwide Revenue")
plt.gca().invert_yaxis()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.hist(marvel_data['Bud­get (mil­lions)'], bins=15, color='orange', edgecolor='black')
plt.xlabel("Budget (in millions)")
plt.ylabel("Number of Movies")
plt.title("Distribution of Marvel Movies Budget")
plt.show()
```

```python
# Convert the 'North America' and 'Other territories' columns to numeric values
marvel_data['North America'] = pd.to_numeric(marvel_data['North America'].str.replace('$', '').str.replace(',', ''), errors='coerce')
marvel_data['Other territories'] = pd.to_numeric(marvel_data['Other territories'].str.replace('$', '').str.replace(',', ''), errors='coerce')

labels = ['North America', 'Other Territories']
sizes = [marvel_data['North America'].sum(), marvel_data['Other territories'].sum()]
colors = ['lightblue', 'lightcoral']

plt.figure(figsize=(7, 7))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors, startangle=140)
plt.title("Revenue Split: North America vs. Other Territories")
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Converting Release Date to Year
marvel_data['Year'] = pd.to_datetime(marvel_data['Release date(United States)'], errors='coerce').dt.year

# Grouping by Year and summing worldwide revenue
yearly_revenue = marvel_data.groupby('Year')['Worldwide'].sum()

plt.figure(figsize=(12, 6))
plt.plot(yearly_revenue.index, yearly_revenue.values, marker='o', linestyle='-', color='b')
plt.xlabel("Year")
plt.ylabel("Total Worldwide Revenue (in billions)")
plt.title("Marvel Movies Revenue Over the Years")
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(marvel_data['Bud­get (mil­lions)'], marvel_data['Worldwide'], color='red', alpha=0.6)
plt.xlabel("Budget (in millions)")
plt.ylabel("Worldwide Revenue")
plt.title("Budget vs. Worldwide Revenue")
plt.grid()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Convert 'Worldwide' column to numeric, forcing errors to NaN
marvel_data['Worldwide'] = pd.to_numeric(marvel_data['Worldwide'].str.replace('$', '').str.replace(',', ''), errors='coerce')

# Get top 10 highest-grossing movies
top_10 = marvel_data.nlargest(10, 'Worldwide')

plt.figure(figsize=(12, 6))
colors = sns.color_palette("coolwarm", len(top_10))  # Gradient colors
plt.barh(top_10['Title'], top_10['Worldwide'], color=colors)
plt.xlabel("Revenue (in billions)")
plt.ylabel("Movies")
plt.title("🌟 Top 10 Marvel Movies by Worldwide Revenue")
plt.gca().invert_yaxis()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
colors = marvel_data['Worldwide'] / marvel_data['Worldwide'].max()  # Normalize colors
plt.scatter(marvel_data['Bud­get (mil­lions)'], marvel_data['Worldwide'], c=colors, cmap='plasma', alpha=0.75, edgecolors="black")
plt.colorbar(label="Revenue Scale")
plt.xlabel("Budget (in millions)")
plt.ylabel("Worldwide Revenue")
plt.title("🔥 Budget vs. Worldwide Revenue (Marvel Movies)")
plt.grid()
plt.show()
```
